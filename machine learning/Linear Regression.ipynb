```python
import os
import numpy
import pylab

print("working directory: {}".format(os.getcwd()))
data = numpy.loadtxt('datasets/dataset.csv', delimiter=',')
# print("data: {}".format(data))
# plot the data
pylab.scatter(data[:, 0], data[:, 1], c='r')
pylab.title('Auto Insurance in Sweden')
pylab.xlabel('total payment for all the claims in thousands of Swedish Kronor')
pylab.ylabel('total number of claims ')
pylab.show()
```

```python
# number of samples
m = len(data)

# get 'm' 1's and de xi's (data[:, 0]), and transpose: [[1,1...1], [x0, x1...xm]] -(T)-> [[1, x0], [1, x1], ...]
X = numpy.array([numpy.ones(m), data[:, 0]]).T

# get the yi's (data[:, 1])
# reshape(n, m) array in to matrix of n rows and m columns. -1 => maximum
# m rows of yi's values: [[y0], [y1] ...]
y = numpy.array(data[:, 1]).reshape(-1, 1)

# compute best-fit: betahat
# betahat = inv(X' * X) * X' * y
betahat = numpy.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
print("betahat: {}".format(betahat))
```

```python
# samples for the regression line, with 2 is enough
num_samples = 2
lin_x = numpy.linspace(0, 130, num_samples)
print(lin_x)
lin_y = numpy.array(betahat[0] + betahat[1] * lin_x)
print(lin_y)
pylab.scatter(data[:, 0], data[:, 1], c='r')
pylab.title('Auto Insurance in Sweden')
pylab.xlabel('total payment for all the claims in thousands of Swedish Kronor')
pylab.ylabel('total number of claims ')
pylab.plot(lin_x, lin_y, c='b')
pylab.show()
```
