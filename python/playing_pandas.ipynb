```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

# Series

```python
ys = [[0,1,2,3,4],[4,3,2,1,0]]
x_ax = [0,1,2,3,4]
for y_ax in ys:
    ts = pd.Series(y_ax,index=x_ax)
    ts.plot(kind='bar', figsize=(15,5))
    plt.show()
```

```python
ts = pd.Series(np.random.randint(0, 10, size=(100,)), index=range(0, 100))
ts.plot(kind='line', figsize=(15,5))
plt.show()
```
